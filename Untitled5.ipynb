```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
from matplotlib.ticker import MaxNLocator

# CSV 読み込み
# ファイル名を適宜変更（例：drive にある場合はパスを指定）
csv_path = "generation_stats_64_128_4000_09_03.csv"
df = pd.read_csv(csv_path)

df["Generation"] = df["Generation"].astype(int)

gen_max = int(df["Generation"].max())
iterations_max = df["Iterations"].max()

# データ量に応じて図サイズを自動調整
# 1世代あたり0.18インチ程度の幅、最低8インチ、最大25インチ
fig_width = min(max(gen_max * 0.18, 8), 25)
fig_height = 6
fig, ax1 = plt.subplots(figsize=(fig_width, fig_height))

# 横軸の範囲を 0 ～ Generation_max + 1 に設定
ax1.set_xlim(0, gen_max + 1)

# 横軸は整数目盛だけにする
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

# 横軸を 10 世代ずつ 白 / 薄灰色で交互に塗り分け
# バンドは最大値+1を含むように、10の倍数まで少し広めに取る
band_max = (( (gen_max + 1) // 10) + 1) * 10
for start in range(0, band_max, 10):
    end = start + 10
    if (start // 10) % 2 == 1:   # 奇数ブロックを薄い灰色
        ax1.axvspan(start, end, facecolor="#e6e6e6", alpha=1.0)
    else:                        # 偶数ブロックを白
        ax1.axvspan(start, end, facecolor="white", alpha=1.0)

# 左軸: ColoringCost（赤・丸）
ax1.plot(
    df["Generation"],
    df["ColoringCost"],
    color="red",
    marker="o",
    label="ColoringCost"
)
ax1.set_xlabel("Generation")
ax1.set_ylabel("ColoringCost", color="red")
ax1.tick_params(axis="y", labelcolor="red")

# 右軸: Iterations（青・×）
ax2 = ax1.twinx()
ax2.plot(
    df["Generation"],
    df["Iterations"],
    color="blue",
    marker="x",
    label="Iterations"
)
ax2.set_ylabel("Iterations", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")

# 右側の縦軸の範囲を Iterations の最大値より少し大きく設定
if iterations_max > 0:
    ax2.set_ylim(0, iterations_max * 1.05)
else:
    # すべて 0 の場合の保険
    ax2.set_ylim(0, 1)

# 右軸の目盛りも整数にする（大きな値でも整数刻み）
ax2.yaxis.set_major_locator(MaxNLocator(integer=True))

# 縦軸は左側のみ自動スケールさせる
ax1.relim()
ax1.autoscale(axis="y")

# Generation が 10 ごとに灰色縦線
for g in range(0, band_max + 1, 10):
    ax1.axvline(x=g, color="gray", linestyle="--", linewidth=0.8)

# グラフタイトル
plt.title("世代ごとの ColoringCost と Iterations の推移")

# レイアウト調整と描画
fig.tight_layout()
plt.show()
```
